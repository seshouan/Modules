## 1. Import Libraries and Dependencies

```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

## 2. Set File Path to CSV using Pathlib

```python
# Create reference to CSV file
# Hint - the file path may require some extra steps..
filepath = Path('../Resources/world_happiness.csv')
```

## 3. Import CSV into Pandas DataFrame

```python
# Import the CSV into a pandas DataFrame
df = pd.read_csv(filepath)
df
```

## 4. Set Index and Sort Index in Ascending Order

```python
# Set new index to Country
df.set_index('Country',inplace=True)

# Sort the new dataframe alphabetically
df.sort_index(inplace=True)
df
```

## 8. Select all the Countries Where GDP is Greater than or Equal to 1.6

```python
# Select all rows where the GDP per capita has a value greater than or equal to 1.6
sample2 = df.loc[df['Economy..GDP.per.Capita.'] >= 1.6]
sample2
```

## 9. Select all the Countries Where the Happiness Score is Greater than or Equal to 5 and GDP is Less than or Equal to 1

```python
# Select all rows that have a happiness score greater than or equal 5 and a GDP score less than or equal 1
sample3 = df.loc[(df['Happiness.Score'] >= 5) & (df['Economy..GDP.per.Capita.'] <= 1)]
sample3
```
