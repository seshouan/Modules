# Alpaca Demo

```python
# Initial imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
# Load .env environment variables
load_dotenv('../../my_env.env')
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv('ALPACA_API_KEY')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')
```

```python
# Verify that Alpaca key and secret were correctly loaded
print(f'Alpaca Key type: {type(alpaca_api_key)}')
print(f'Alpaca Secret Key type: {type(alpaca_secret_key)}')
```

```python
# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
today = pd.Timestamp("2020-07-14", tz="America/New_York").isoformat()
```

```python
# Set the tickers
tickers = ['FB','TWTR']
```

```python
# Set timeframe to one day ("1Day") for the Alpaca API
timeframe = "1Day"
```

```python
# Get current closing prices for FB and TWTR
df_portfolio = alpaca.get_bars(
    tickers,
    timeframe,
    start = today,
    end = today
).df

# Display sample data
df_portfolio
```

```python
# Format start and end dates as ISO format for one year period
start = pd.Timestamp("2019-07-14", tz="America/New_York").isoformat()
end = pd.Timestamp("2020-07-14", tz="America/New_York").isoformat()
```

```python
# Get closing prices for FB and TWTR from the last year
df_portfolio = alpaca.get_bars(
    tickers,
    timeframe,
    start = start,
    end = end
).df

# Display sample data
df_portfolio.head(10)
```

```python
# Reorganize the DataFrame

# Separate ticker data
FB = df_portfolio[df_portfolio['symbol']=='FB'].drop('symbol',axis=1)
TWTR = df_portfolio[df_portfolio['symbol']=='TWTR'].drop('symbol',axis=1)
# Concatenate the ticker DataFrames
df_portfolio = pd.concat([FB,TWTR],axis=1,keys=['FB','TWTR'])
# Display sample data
df_portfolio
```

```python
df_portfolio.tail(10)
```

```python
# Create and empty DataFrame for closing prices
combined_df = pd.DataFrame()
# Fetch the closing prices of FB and TWTR
combined_df['FB'] = df_portfolio['FB']['close']
combined_df['TWTR'] = df_portfolio['TWTR']['close']
# Drop the time component of the date
combined_df.index = combined_df.index.date
# Display sample data
combined_df.head(10)
```

```python
# Compute daily returns
combined_returns = combined_df.pct_change().dropna()
# Display sample data
combined_returns.head()
```

```python
# Plot daily returns
combined_returns.plot()
```
