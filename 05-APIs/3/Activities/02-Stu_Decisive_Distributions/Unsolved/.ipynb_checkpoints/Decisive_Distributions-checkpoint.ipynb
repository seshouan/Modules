## Decisive Distributions

### Import Dependencies

```python
# Import libraries and dependencies
import os
import pandas as pd
import alpaca_trade_api as tradeapi

# Load .env enviroment variables
from dotenv import load_dotenv

%matplotlib inline
```

```python
load_dotenv('../../../../my_env.env')
```

```python
# Setup Alpaca API environment
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

### Read in stock data using Alpaca API

```python
# Set timeframe to "1Day"
timeframe = "1Day"

# Set start and end datetimes of 1 year, between now and 365 days ago.
start_date = pd.Timestamp("2022-07-12", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2023-07-12", tz="America/New_York").isoformat()

# Set the stock tickers
tickers = ['SPY','LUV','DIS','AAPL','SBUX','WORK']

# Get 1 year's worth of historical data for all stocks
# HINT: Set "limit" to 10000 so all ticker rows are captured from get_bars()
df_portfolio = api.get_bars(
    tickers,
    timeframe,
    start = start_date,
    end = end_date,
    limit = 10000
).df

# Display sample data
df_portfolio.head()
```

```python
# Reorganize the DataFrame
# Separate ticker data
SPY = df_portfolio[df_portfolio['symbol']=='SPY'].drop('symbol', axis=1)
LUV = df_portfolio[df_portfolio['symbol']=='LUV'].drop('symbol', axis=1)
DIS = df_portfolio[df_portfolio['symbol']=='DIS'].drop('symbol', axis=1)
AAPL = df_portfolio[df_portfolio['symbol']=='AAPL'].drop('symbol', axis=1)
SBUX = df_portfolio[df_portfolio['symbol']=='SBUX'].drop('symbol', axis=1)
WORK = df_portfolio[df_portfolio['symbol']=='WORK'].drop('symbol', axis=1)

# Concatenate the ticker DataFrames
df_portfolio = pd.concat([SPY, LUV, DIS, AAPL, SBUX, WORK], axis=1, keys=["SPY", "LUV", "DIS", "AAPL", "SBUX", "WORK"])
# Display sample data
df_portfolio.head()
```

### Create a new DataFrame and store the closing prices of each stock.

```python
# Create and empty DataFrame for closing prices
df_closing_prices = pd.DataFrame()

# Fetch the closing prices for all the tickers
df_closing_prices['SPY'] = df_portfolio['SPY']['close']
df_closing_prices['LUV'] = df_portfolio['LUV']['close']
df_closing_prices['DIS'] = df_portfolio['DIS']['close']
df_closing_prices['AAPL'] = df_portfolio['AAPL']['close']
df_closing_prices['SBUX'] = df_portfolio['SBUX']['close']
df_closing_prices['WORK'] = df_portfolio['WORK']['close']

# Drop the time component of the date
df_closing_prices.index = df_closing_prices.index.date

# Display sample data
df_closing_prices.head()
```

### Calculate the daily returns for each stock using the `pct_change()` function

```python
# Drop 'WORK'
df_closing_prices.drop('WORK',axis=1,inplace=True)

# Compute daily returns
df_daily_returns = df_closing_prices.pct_change().dropna()

# Display sample data
df_daily_returns.head()
```

### Plot Distributions

```python
# Visualize the distribution of daily returns across all stocks using a histogram plot
# Hint: To make the plot easier to read, set the alpha arguement to 0.5
df_daily_returns.plot.hist(alpha=0.5)
```

```python
# Visualize the distribution of daily returns across all stocks using a density plot
df_daily_returns.plot.density(alpha=0.5)
```
